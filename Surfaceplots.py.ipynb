## Surface Plots
- surface plots are used to 
- visualise loss function in machine learning and deep learning
- visualise state or state value functions in reinforcment learning

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7])
a,b = np.meshgrid(a,b)
```

```python
print(a)
```

```python
print(b)
```

```python
##In this a along row is repeated and b is repeated along column b.length() no of times 
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
#axes.plot_surface(a,b,a**2+b**2,cmap='rainbow')
plt.show()
##THIS IS GIVING ME 3D AXIS
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
                #xaxis,yaxix,zaxis
axes.plot_surface(a,b,a**2+b**2,cmap='rainbow')
plt.show()
##RED MEANS REGION OF HIFH VALUE BLUE MEANS REGION OF LOW VALUE
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
                #xaxis,yaxix,zaxis
axes.plot_surface(a,b,a*b,cmap='rainbow')
plt.show()
```

```python
a = np.arange(-4,4,0.02)
b = a

a,b = np.meshgrid(a,b)

print(a)
print(b)
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
                #xaxis,yaxix,zaxis
axes.plot_surface(a,b,a**2+b**2,cmap='rainbow')
plt.show()
```

```python
###IN CONTOUR PLANE WE CUT THIS SURFACE WITH PARLLEL PALIN
```

```python
fig = plt.figure()
axes.contour(a,b,a**2+b**2,cmap='rainbow')
plt.show()
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.contour(a,b,a**2+b**2,cmap='rainbow')
plt.show()
```
